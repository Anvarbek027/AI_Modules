# **1- MASALA:** **Vacuum - Cleaner World agentini yaratish. (2 ta xona uchun)**



---
##**Maqsad:**
##2 ta xonali (A va B) muhitda tozalikni saqlash.

## **Vazifa:**

## Agent hozir qaysi xonada ekanligini va xonaning holatini (toza yoki kir) aniqlashi kerak.

* ## *Agar xona "Kir" bo'lsa -> "Tozalamoq".*

* ## *Agar xona "Toza" bo'lsa va A da bo'lsa -> "O'ngga yurish".*

* ## *Agar xona "Toza" bo'lsa va B da bo'lsa -> "Chapga yurish".*


#**‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>xonalar = {"A": "iflos", "B": "toza"}</code>

* ### <code>agent_joylashuvi = "A"</code>

##***Klasslar:***

* ### <code>Muhit</code>
  * ### <code>perseptsiya_ol()</code>
  * ### <code>harakat_qollash(harakat)</code>
* ### <code>RefleksAgent</code>
  * ### <code>qaror_qabul_qil(perseptsiya)</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas
 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.

```python
# Kodni shu yerga yozing
class Muhit:
    def __init__(self, xonalar, agent_joylashuvi):
        self.xonalar = xonalar
        self.agent_joylashuvi = agent_joylashuvi


    def perseptsiya_ol(self):
        holat = self.xonalar[self.agent_joylashuvi]
        return (self.agent_joylashuvi, holat)

    def harakat_qollash(self, harakat):
        if harakat == "Tozalamoq":
            self.xonalar[self.agent_joylashuvi] = "toza"
            print(f"{self.agent_joylashuvi} xonasi tozalandi.")

        elif harakat == "O'ngga yurish":
            self.agent_joylashuvi = "B"
            print("Agent B xonasiga o‚Äòtdi.")

        elif harakat == "Chapga yurish":
            self.agent_joylashuvi = "A"
            print("Agent A xonasiga o‚Äòtdi.")



class RefleksAgent:
    def qaror_qabul_qil(self, perseptsiya):
        joylashuv, holat = perseptsiya

        if holat == "iflos":
            return "Tozalamoq"
        elif joylashuv == "A":
            return "O'ngga yurish"
        elif joylashuv == "B":
            return "Chapga yurish"



xonalar = {"A": "iflos", "B": "toza"}
agent_joylashuvi = "A"

muhit = Muhit(xonalar, agent_joylashuvi)
agent = RefleksAgent()


for i in range(20):
    print(f"\n{i+1}-qadam")
    perseptsiya = muhit.perseptsiya_ol()
    harakat = agent.qaror_qabul_qil(perseptsiya)
    print("Agent qarori:", harakat)
    muhit.harakat_qollash(harakat)
```

# **2-MASALA: Haroratni nazorat qiluvchi agentni yaratish.**


---


## **Maqsad:**

##Sonli ma‚Äôlumot (harorat) asosida refleks qaror chiqarishni o‚Äòrganish.

## **Vazifa:**

## Xona haroratiga qarab isitish yoki sovitish qurilmasini boshqarish.

* ### harorat < 18 ‚Üí **ISITISH_YOQ**

* ### harorat > 26 ‚Üí **SOVITISH_YOQ**

* ###  Aks holda ‚Üí **KUTISH**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>harorat</code> (10 - 35 oralig'ida)



##***Klasslar:***

```python
# Kodni shu yerga yozing
import random
class HaroratMuhiti:
    def __init__(self, harorat):
        self.harorat = harorat

    def get_harorat(self):
        return self.harorat


class HaroratAgenti:
    def qaror_chiqarish(self, harorat):
        if harorat < 18:
            return "ISITISH_YOQ"
        elif harorat > 26:
            return "SOVITISH_YOQ"
        else:
            return "KUTISH"

for i in range(20):
    harorat = random.randint(10, 35)
    muhit = HaroratMuhiti(harorat)
    agent = HaroratAgenti()
    qaror = agent.qaror_chiqarish(muhit.get_harorat())
    print(f"{i+1}-ma'lumot: Harorat = {muhit.get_harorat()}¬∞C ‚Üí Agent qarori: {qaror}")
```

# **3-MASALA: Aqlli sug‚Äòorish agentini yaratish**

---


## **Maqsad:**

## Agentga mantiqiy va (and) operatori hamda inkor (not) tushunchasini qo‚Äòllash.

## **Vazifa:**

## Tuproq namligi va yomg‚Äòir bor-yo‚Äòqligiga qarab suv nasosini boshqarish.

* ### yomgir = True ‚Üí **NASOSNI_OCHIR**

* ### namlik < 30 va yomgir = False ‚Üí **NASOSNI_YOQ**

* ###  namlik >= 30 va yomgir == False ‚Üí NASOS_OCH ‚Üí **NASOSNI_OCHIR**
* ### Aks holda ‚Üí **TEKSHIRISH_REJIMI**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>namlik</code> (0-100 oralig'idagi son)
* ### <code>yomgir</code> (True yoki False)



##***Klasslar:***

* ### <code>BogMuhiti</code>

* ### <code>SugorishAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas


 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.


```python
# Kodni shu yerga yozing
import random


class BogMuhiti:
    def __init__(self, namlik, yomgir):
        self.namlik = namlik
        self.yomgir = yomgir
    def get_namlik(self):
        return self.namlik

    def get_yomgir(self):
        return self.yomgir

class SugorishAgenti:
    def qaror_chiqarish(self, namlik, yomgir):


        if yomgir == True:
            return "NASOSNI_OCHIR"


        elif namlik < 30 and not yomgir:
            return "NASOSNI_YOQ"


        elif namlik >= 30 and not yomgir:
            return "NASOSNI_OCHIR"


        else:
            return "TEKSHIRISH_REJIMI"



for i in range(20):
    namlik = random.randint(0, 100)
    yomgir = random.choice([True, False])

    muhit = BogMuhiti(namlik, yomgir)
    agent = SugorishAgenti()

    qaror = agent.qaror_chiqarish(
        muhit.get_namlik(),
        muhit.get_yomgir()
    )

    print(f"{i+1}-holat: Namlik = {namlik}%, Yomgir = {yomgir} ‚Üí Qaror: {qaror}")
```

# **4-MASALA: Xavfsizlik qorovuli agentini yaratish.**


---


## **Maqsad:**

## Agent ikki xil sensordan kelgan ma'lumotlarni birlashtirib qaror qabul qilish.

## **Vazifa:**

## Harakat sensori va eshik qulfi holatiga qarab signal chalish.

* ### harakat bor va qulf ochiq bo'lsa  ‚Üí **SIGNAL_CHALISH**

* ### harakat bor va qulf yopiq bo'lsa ‚Üí **SALOMLASHISH**

* ###  harakat yo'q ‚Üí **JIM_TURISH**
* ### Aks holda ‚Üí **KUZATISH**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>harakat</code> (True yoki False)
* ### <code>qulf_ochiq</code> (True yoki False)



##***Klasslar:***

* ### <code>UyMuhiti</code>

* ### <code>QorovulAgenti</code>


##***Simulyatsiya:***
* ### 20 tadan kam emas

 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.



```python
# Kodni shu yerga yozing
import random

class UyMuhit:
    def __init__(self):
        self.harakat = random.choice([True, False])
        self.qulf_ochiq = random.choice([True, False])

    def yangilash(self):
        self.harakat = random.choice([True, False])
        self.qulf_ochiq = random.choice([True, False])


class GorovulAgenti:
    def qaror_qabul(self, harakat, qulf_ochiq):
        if harakat and qulf_ochiq:
            return "SIGNAL_CHALISH "
        elif harakat and not qulf_ochiq:
            return "SALOMLASHISH"
        elif not harakat:
            return "JIM_TURISH"
        else:
            return "KUZATISH "


muhit = UyMuhit()
agent = GorovulAgenti()

for i in range(20):
    print(f"\n--- {i+1}-holat ---")
    print("Harakat:", muhit.harakat)
    print("Qulf ochiq:", muhit.qulf_ochiq)

    qaror = agent.qaror_qabul(muhit.harakat, muhit.qulf_ochiq)
    print("Agent qarori:", qaror)

    muhit.yangilash()
```

```python
# Kodni shu yerga yozing
import random

class BinoMuhiti:
    def __init__(self):
        self.lift_gavati = random.randint(1, 9)
        self.chagiruv_gavati = random.randint(1, 9)

    def yangilash(self):
        self.lift_gavati = random.randint(1, 9)
        self.chagiruv_gavati = random.randint(1, 9)


class LiftAgenti:
    def qaror_qabul(self, lift_gavati, chagiruv_gavati):
        if lift_gavati < chagiruv_gavati:
            return "YUQORIGA ‚¨Ü"
        elif lift_gavati > chagiruv_gavati:
            return "PASTGA ‚¨á"
        elif lift_gavati == chagiruv_gavati:
            return "ESHIKNI_OCH "
        else:
            return "KUTISH "


muhit = BinoMuhiti()
agent = LiftAgenti()

for i in range(20):
    print(f"\n--- {i+1}-holat ---")
    print("Lift joylashgan qavat:", muhit.lift_gavati)
    print("Chaqirilgan qavat:", muhit.chagiruv_gavati)

    qaror = agent.qaror_qabul(muhit.lift_gavati, muhit.chagiruv_gavati)
    print("Agent qarori:", qaror)

    muhit.yangilash()
```

# **8-MASALA: Spam aniqlovchi agent yaratish.**

---

## **Maqsad:**

## Matnli ma'lumotlar (String) ichidan kalit so‚Äòzlarni qidirish va saralashni o‚Äòrganish.

## **Vazifa:**

## Kelgan xabarning tarkibida shubhali so‚Äòzlar bor-yo‚Äòqligiga qarab uni tegishli papkaga joylashtirish.

* ### xabar tarkibida "yutuq" yoki "tekin" so‚Äòzlari qatnashsa ‚Üí **SPAM_PAPKA**
* ### xabar tarkibida "hisobot" yoki "ish" so‚Äòzlari qatnashsa ‚Üí **MUHIM_PAPKA**
* Aks holda ‚Üí **KIRUVCHI_PAPKA**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>xabar</code> (String - matn ko‚Äòrinishida)

##***Klasslar:***

* ### <code>PochtaMuhiti</code>

* ### <code>SpamAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas

 ## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.


```python
# Kodni shu yerga yozing
import random

class PochtaMuhiti:
    def __init__(self):
        self.xabarlar = [
            "Siz katta yutuq qo'lga kiritdingiz",
            "Bugun tekin sovg'a bor",
            "Ish bo'yicha yangi taklif",
            "Oylik hisobotni yuboring",
            "Salom, uchrashamizmi?",
            "Tezkor yutuq sizni kutmoqda",
            "Yangi ish loyihasi boshlandi",
            "Hisobot tayyormi?",
            "Oddiy xabar matni",
            "Tekin bonus faqat bugun",
            "Rahmat sizga",
            "Ish yuzasidan suhbat",
            "Yutuq uchun ro'yxatdan o'ting",
            "Hisobot faylini tekshiring",
            "Do'stlik taklifi",
            "Tez pul va tekin imkoniyat",
            "Ish rejasini ko'rib chiqing",
            "Oddiy salomlashuv",
            "Yutuq va sovrinlar",
            "Hisobot topshirish muddati"
        ]
        self.xabar = random.choice(self.xabarlar)

    def yangilash(self):
        self.xabar = random.choice(self.xabarlar)


class SpamAgenti:
    def qaror_qabul(self, xabar):
        matn = xabar.lower()
        if "yutuq" in matn or "tekin" in matn:
            return "SPAM_PAPKA "
        elif "hisobot" in matn or "ish" in matn:
            return "MUHIM_PAPKA "
        else:
            return "KIRUVCHI_PAPKA "


muhit = PochtaMuhiti()
agent = SpamAgenti()
for i in range(20):
    print(f"\n--- {i+1}-xabar ---")
    print("Xabar:", muhit.xabar)

    qaror = agent.qaror_qabul(muhit.xabar)
    print("Joylashtirildi:", qaror)

    muhit.yangilash()
```

## **10-MASALA: Aqlli muzlatkich agentini yaratish.**

---

## **Maqsad:**

## Bir nechta (3 ta) sensordan kelgan ma'lumotlarni ustuvorlik bo‚Äòyicha tahlil qilish.

## **Vazifa:**

## Muzlatkich eshigi, ichki harorat va mahsulot miqdoriga qarab, mos xabarni chiqarish.

* ### **SIGNAL_CHALISH** (Eng muhimi)
* ### **SOVITISHNI_KUCHAYTIRISH**
* ### **NORMAL_REJIM**
* ### Aks holda ‚Üí **BUYURTMA_BERISH**
* ### Aks holda ‚Üí **NORMAL_HOLAT**

## **‚öôÔ∏è Bajarish tartibi**

##***Ma'lumotlar:***

* ### <code>eshik_ochiq</code> (True yoki False)

* ### <code>harorat</code> (butun son, masalan: 4, 8, -2)

* ### <code>mahsulot_soni</code> (butun son)

##***Klasslar:***

* ### <code>OshxonaMuhiti</code>

* ### <code>MuzlatkichAgenti</code>

##***Simulyatsiya:***
* ### 20 tadan kam emas

## **üö´Eslatma:** Masaladagi o'zgaruvchilar, ma'lumotlar, klasslar bajarish tartibida keltirilgan nomlar bilan nomlansin.


```python
# Kodni shu yerga yozing
import random

class OshxonaMuhiti:
    def __init__(self):
        self.eshik_ochig = random.choice([True, False])
        self.harorat = random.randint(-5, 15)
        self.mahsulot_soni = random.randint(0, 10)

    def yangilash(self):
        self.eshik_ochig = random.choice([True, False])
        self.harorat = random.randint(-5, 15)
        self.mahsulot_soni = random.randint(0, 10)


class MuzlatkichAgenti:
    def qaror_qabul(self, eshik_ochig, harorat, mahsulot_soni):
        if eshik_ochig:
            return "SIGNAL_CHALISH "
        elif harorat > 8:
            return "SOVITISHNI_KUCHAYTIRISH "
        elif mahsulot_soni <= 1:
            return "BUYURTMA_BERISH "
        elif 0 <= harorat <= 8:
            return "NORMAL_REJIM "
        else:
            return "NORMAL_HOLAT "


muhit = OshxonaMuhiti()
agent = MuzlatkichAgenti()

for i in range(20):
    print(f"\n--- {i+1}-holat ---")
    print("Eshik ochiq:", muhit.eshik_ochig)
    print("Harorat:", muhit.harorat)
    print("Mahsulot soni:", muhit.mahsulot_soni)

    qaror = agent.qaror_qabul(
        muhit.eshik_ochig,
        muhit.harorat,
        muhit.mahsulot_soni
    )
    print("Agent qarori:", qaror)

    muhit.yangilash()
```
